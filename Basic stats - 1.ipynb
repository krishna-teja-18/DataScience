# Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data Visualisation

# Plotting histogram for Volume
plt.figure(figsize=(8, 5))
sns.histplot(data["Volume"],bins=30,kde='True')
plt.title("Distribution of Volume")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()
```

# Volume: 

Skeweness: Right-skeweness or Positive Skewed 

Outliers: Some transactions have a very high number of units sold, which are rare cases.

```python
#Histogram for Avg Price

plt.figure(figsize=(8, 5))
sns.histplot(data["Avg Price"],bins=30,kde=True)
plt.title("Distribution of Avg Price")
plt.xlabel("Avg Price")
plt.ylabel("Frequency")
plt.show()
```

# Avg Price:

Skewness: Highly right-skewed (most products have low prices, but some are very expensive).

Outliers: Expensive products create extreme values.

```python
#Histogram for Total Sales Value

plt.figure(figsize=(8, 5))
sns.histplot(data["Total Sales Value"], bins=30, kde=True)
plt.title("Distribution of Total Sales Value")
plt.xlabel("Total Sales Value")
plt.ylabel("Frequency")
plt.show()
```

# Total Sales Value:

Skewness: Right-skewed.

Outliers: Some transactions generate very high sales, affecting the overall average.

```python
#Histogram for Discount Rate (%)

plt.figure(figsize=(8, 5))
sns.histplot(data["Discount Rate (%)"], bins=30, kde=True)
plt.title("Distribution of Discount Rate (%)")
plt.xlabel("Discount Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

Skewness: Slightly skewed but close to normal.

Outliers: Some transactions have unusually high or low discount rates.


```python
#Histogram for Discount Amount

plt.figure(figsize=(8, 5))
sns.histplot(data["Discount Amount"], bins=30, kde=True)
plt.title("Distribution of Discount Amount")
plt.xlabel("Discount Amount")
plt.ylabel("Frequency")
plt.show()
```

Skewness: Right-skewed (most discounts are small, but a few are very large).

Outliers: A few transactions receive significantly higher discounts.

```python
#Histogram for Net Sales Value

plt.figure(figsize=(8, 5))
sns.histplot(data["Net Sales Value"], bins=30, kde=True)
plt.title("Distribution of Net Sales Value")
plt.xlabel("Net Sales Value")
plt.ylabel("Frequency")
plt.show()
```

Skewness: Right-skewed 

Outliers: Some sales contribute significantly more than others, making them stand out.

# Bar Charts

```python
# Bar Chart for Day
plt.figure(figsize=(10, 5))
sns.countplot(y=data["Day"], order=data["Day"].value_counts().index, palette="viridis")
plt.title("Frequency of Day")
plt.xlabel("Count")
plt.ylabel("Day")
plt.show()
```

```python
#Bar chart for SKU

plt.figure(figsize=(10, 5))
sns.countplot(y=data["SKU"], order=data["SKU"].value_counts().index[:10], palette="viridis")  # Top 10 only
plt.title("Top 10 Most Frequent SKUs")
plt.xlabel("Count")
plt.ylabel("SKU")
plt.show()
```

```python
# Bar Chart for city

plt.figure(figsize=(10, 5))
sns.countplot(y=data["City"], order=data["City"].value_counts().index[:10], palette="viridis")  # Top 10 only
plt.title("Top 10 Cities by Frequency")
plt.xlabel("Count")
plt.ylabel("City")
plt.show()
```

```python
#Bar chart for Business Unit

plt.figure(figsize=(10, 5))
sns.countplot(y=data["BU"], order=data["BU"].value_counts().index, palette="viridis")
plt.title("Frequency of Business Units (BU)")
plt.xlabel("Count")
plt.ylabel("BU")
plt.show()
```

```python
# Bar chart for Brand

plt.figure(figsize=(10, 5))
sns.countplot(y=data["Brand"], order=data["Brand"].value_counts().index[:10], palette="viridis")  # Top 10 only
plt.title("Top 10 Most Frequent Brands")
plt.xlabel("Count")
plt.ylabel("Brand")
plt.show()
```

```python
# Bar chart for model

plt.figure(figsize=(10, 5))
sns.countplot(y=data["Model"], order=data["Model"].value_counts().index[:10], palette="viridis")  # Top 10 only
plt.title("Top 10 Most Frequent Models")
plt.xlabel("Count")
plt.ylabel("Model")
plt.show()
```

The dataset shows that sales are not evenly distributed across different categories.

Days: Some days have significantly higher sales, indicating peak business periods, while others have lower activity. Businesses can use this insight to plan promotions on slower days to boost sales.

SKUs (Products): A few products contribute to most sales, while others sell rarely. This suggests that high-demand products should be prioritized for stock replenishment, while slow-moving items may need discounts or discontinuation.

Cities: Certain cities generate more sales, indicating strong demand in specific regions. Expanding business in these high-performing cities or improving marketing in weaker cities can enhance growth.

Brands: Some brands dominate sales, showing customer preference, while others lag behind. Companies can focus marketing efforts on popular brands and assess whether underperforming brands need better promotion or pricing adjustments.

Business Units (BU): Some business units contribute significantly more sales than others. Analyzing what makes top-performing units successful can help replicate success strategies in weaker units.

Models: Certain product models sell more frequently, showing market preference, while others struggle. Businesses can optimize their product lineup based on this trend.

```python
# Standardization (Z-score normalization)
standardized_df = (numerical_columns - numerical_columns.mean()) / numerical_columns.std()

# Combine original and standardized data for comparison
comparison_df = pd.concat([numerical_columns, standardized_df.add_suffix('_standardized')], axis=1)

print(comparison_df.iloc[0])
```

```python
# Plot before and after distributions
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Before Standardization
sns.histplot(numerical_columns, kde=True, ax=axes[0])
axes[0].set_title("Before Standardization")

# After Standardization
sns.histplot(standardized_df, kde=True, ax=axes[1])
axes[1].set_title("After Standardization")

plt.show()
```
